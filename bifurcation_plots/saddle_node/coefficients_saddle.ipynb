```python
from All_Functions_saddle import *
import sympy as sp
import numpy as np
import copy
import scipy.stats as scp
import math
import matplotlib.pyplot as plt
```

```python
N_PC = 2                    # Truncation degree of Polynomial Chaos expansion

mu_low = sp.Rational(-1,2)       # left extremum of the interval
mu_up = sp.Rational(1,1)        # right extremum of the interval

# Mean and standard deviation (note: sigma must be scaled by sqrt(3) and then squared to obtain the variance)
N_sample_mu = 1
h = sp.Rational(mu_up - mu_low, N_sample_mu)
mu = np.arange(mu_low, mu_up + h, h)

sigma_low = sp.Rational(0,1)
sigma_up = sp.Rational(1,10)

N_sample_sigma = 1
k = sp.Rational(sigma_up - sigma_low, N_sample_sigma)
sigma = np.arange(sigma_low, sigma_up + k, k)

solutions = multiple_systems_solver(mu, sigma, N_PC, number_of_solutions = True)
print(solutions)
```

```python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
matplotlib.use('TkAgg')  # o 'Qt5Agg', se hai Qt installato

# Supponiamo che il tuo array si chiami "data"
# Ecco un esempio di come fare il flatten e plot

# 1. Estrai tutti i punti da sotto-liste non vuote
points = []
for sublist in solutions:
    for point in sublist:
        if isinstance(point, np.ndarray) and point.shape == (3,):
            points.append(point)

# 2. Converti in array numpy
points = np.array(points)

# 3. Separa x, y, z
x = points[:, 0]
y = points[:, 1]
z = points[:, 2]

# 4. Plot 3D con matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', marker='o')

# Label e visualizzazione
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("Scatter 3D dei punti")
plt.show()
```
