```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "text.latex.preamble": r"\usepackage{lmodern}",
    "font.family": "lmodern",

    "font.size": 25,          # Base font size
    "axes.labelsize": 28,     # Axis labels
    "axes.titlesize": 30,     # Titles (if used)
    "xtick.labelsize": 25,
    "ytick.labelsize": 25,
    "legend.fontsize": 25,

    "axes.linewidth": 1.2,
    "grid.linewidth": 0.8,
})

x_mu = np.linspace(-1, 2, 2000)
idx_0 = np.argmax(x_mu >= 0)

fig, ax = plt.subplots(
    figsize=(8, 6),  # inches
    dpi=250           # exact output size
)

color = "#f79a25"

idx_0 = np.argmax(x_mu >= 0)

ax.plot(x_mu[idx_0:], np.sqrt(x_mu[idx_0:]), color=color, linestyle="-", label="Stable branch")
ax.plot(x_mu[idx_0:], -np.sqrt(x_mu[idx_0:]), color=color, linestyle="-")
ax.plot(x_mu[idx_0:], [0 for i in range(len(x_mu[idx_0:]))], color=color, linestyle=":", label="Unstable branch")
ax.plot(x_mu[:idx_0], [0 for i in range(len(x_mu[:idx_0]))], color=color, linestyle="-")

ax.set_xlabel(r"$\mu$")
ax.set_ylabel(r"$u$")

ax.set_xlim(-1, 2)
ax.set_ylim(-2, 2)

ax.grid(True)

ax.legend(
    loc="lower left",
    frameon=True,
    borderpad=0.4,
    labelspacing=0.3,
    handlelength=1.4
)
ax.set_title(r"Pitchfork bifurcation", pad=12)
fig.tight_layout()
fig.savefig("Plots/Pitchfork.pdf")
fig.savefig("Plots/Pitchfork.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
    'font.size': 35,           # Dimensione generale del font
    'axes.titlesize': 35,      # Titolo dell'asse
    'axes.labelsize': 35,      # Etichette degli assi
    'legend.fontsize': 35,     # Legenda
    'xtick.labelsize': 35,     # Etichette asse x
    'ytick.labelsize': 35      # Etichette asse y
})

plt.rcParams['text.latex.preamble']=r"\usepackage{lmodern}"
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 25
plt.rcParams['font.family'] = 'lmodern'

x_mu = np.linspace(-2, 1, 1000)

fig, ax = plt.subplots(figsize=(10, 5))

idx_0 = np.argmax(x_mu >= 0)


ax.plot(x_mu[:idx_0], [0 for i in range(len(x_mu[:idx_0]))], color="red", linestyle="-", label="Stable branch")
ax.plot(x_mu[:idx_0], np.sqrt(-x_mu[:idx_0]), color="red", linestyle=":", label="Unstable branch")
ax.plot(x_mu[:idx_0], -np.sqrt(-x_mu[:idx_0]), color="red", linestyle=":")
ax.plot(x_mu[idx_0:], [0 for i in range(len(x_mu[idx_0:]))], color="red", linestyle=":")


# Etichette degli assi con dimensione del font
ax.set_xlabel(r"$\mu$")  
ax.set_ylabel(r"$u$")


# Limiti e griglia
ax.set_xlim([-2, 1])
ax.grid()

ax.legend(loc='lower center', bbox_to_anchor=(0.5, 1), ncol=3)
# Legenda


# Mostra il grafico
fig.savefig(f"Plots/Pitchfork_rev.pdf", bbox_inches="tight")
```
