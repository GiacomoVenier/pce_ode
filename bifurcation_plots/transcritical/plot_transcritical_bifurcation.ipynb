```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "text.latex.preamble": r"\usepackage{lmodern}",
    "font.family": "lmodern",

    "font.size": 25,          # Base font size
    "axes.labelsize": 28,     # Axis labels
    "axes.titlesize": 30,     # Titles (if used)
    "xtick.labelsize": 25,
    "ytick.labelsize": 25,
    "legend.fontsize": 25,

    "axes.linewidth": 1.2,
    "grid.linewidth": 0.8,
})

x_mu = np.linspace(-1, 2, 2000)
idx_0 = np.argmax(x_mu >= 0)



x_mu = np.linspace(-1, 1, 2000)
idx_0 = np.argmax(x_mu >= 0)

fig, ax = plt.subplots(
    figsize=(8, 6),  # inches
    dpi=250           # exact output size
)

color = "#f79a25"

ax.plot(
    x_mu[idx_0:], x_mu[idx_0:],
    color=color, linestyle="-",
    linewidth=2.5,
    label="Stable branch"
)

ax.plot(
    x_mu[:idx_0], x_mu[:idx_0],
    color=color, linestyle=":",
    linewidth=2.5,
    label="Unstable branch"
)

ax.plot(
    x_mu[:idx_0], np.zeros_like(x_mu[:idx_0]),
    color=color, linestyle="-",
    linewidth=2.5
)

ax.plot(
    x_mu[idx_0:], np.zeros_like(x_mu[idx_0:]),
    color=color, linestyle=":",
    linewidth=2.5
)

ax.set_xlabel(r"$\mu$")
ax.set_ylabel(r"$u$")

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)

ax.grid(True)

ax.legend(
    loc="lower right",
    frameon=True,
    borderpad=0.4,
    labelspacing=0.3,
    handlelength=1.4
)
ax.set_title(r"Transcritical bifurcation", pad=12)

fig.tight_layout()
fig.savefig("Plots/Transcritical.pdf")
fig.savefig("Plots/Transcritical.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


plt.rcParams.update({
    'font.size': 35,           # Dimensione generale del font
    'axes.titlesize': 35,      # Titolo dell'asse
    'axes.labelsize': 35,      # Etichette degli assi
    'legend.fontsize': 35,     # Legenda
    'xtick.labelsize': 35,     # Etichette asse x
    'ytick.labelsize': 35      # Etichette asse y
})

plt.rcParams['text.latex.preamble']=r"\usepackage{lmodern}"
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 30
plt.rcParams['font.family'] = 'lmodern'


x_mu = np.linspace(-1, 1, 2000)

fig, ax = plt.subplots(figsize=(10, 5))

idx_0 = np.argmax(x_mu >= 0)

ax.plot(x_mu[idx_0:], -x_mu[idx_0:], color="red", linestyle="-", label="Stable branch")
ax.plot(x_mu[:idx_0], -x_mu[:idx_0], color="red", linestyle=":", label="Unstable branch")
ax.plot(x_mu[:idx_0], [0 for i in range (len(x_mu[:idx_0]))], color="red", linestyle="-")
ax.plot(x_mu[idx_0:], [0 for i in range (len(x_mu[:idx_0]))], color="red", linestyle=":")

# Etichette degli assi con dimensione del font
ax.set_xlabel(r"$\mu$")  
ax.set_ylabel(r"$u$")

# Impostazioni degli assi
#ax.axhline(0, color='black', linewidth=1.3)  # Linea asse x
#ax.axvline(0, color='black', linewidth=1.3)  # Linea asse y

# Limiti e griglia
ax.grid()

# Legenda
ax.legend(loc='lower center', bbox_to_anchor=(0.5, 1), ncol=3)

# Mostra il grafico
fig.savefig(f"Plots/Transcritical_rev.pdf", bbox_inches="tight")
```
